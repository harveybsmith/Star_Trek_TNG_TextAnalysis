```python
import pandas as pd
import json
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv('TNG.csv')
```

```python
df
```

```python
#Lets group the entire corpus of each major character into a dictionary
main_chars = ['PICARD', 'RIKER','DATA', 'LAFORGE' ,'WORF','CRUSHER','TROI', 'WESLEY', 'Q','GUINAN']
```

```python
# Create empty dictionary and apply boolean filter for characters from out list and the .values creates array
char_corpuses = {}
for char in main_chars:
    char_corpuses[char] = df[df.Character == char]['free_dialogue'].values
```

```python
char_corps = {char: "".join(corpus) for char, corpus in char_corpuses.items()}
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
for char, text in char_corps.items():
    char_corps[char] = sent_tokenize(text)
```

```python
sid_obj = SentimentIntensityAnalyzer()
sentiments = {char: sid_obj.polarity_scores(corpus) for char, corpus in char_corps.items()}
   
```

```python
sentiments
```

```python
# save model for later
# Serializing json  
json_object = json.dumps(sentiments, indent = 4) 
  
# Writing to sample.json 
with open("sentiments.json", "w") as outfile: 
    outfile.write(json_object)
```

Runtime was 1.5 hours

```python
for character, score in sentiments.items():
    print("{} was rated as ".format(character), score['neg']*100, "% Negative") 
    print("{} was rated as ".format(character), score['neu']*100, "% Neutral") 
    print("{} was rated as ".format(character), score['pos']*100, "% Positive")
```

## Removing stopwords

```python
from nltk.corpus import stopwords
stoplist = stopwords.words('english')

# Create columns of stopwords removed
df['no_stop_words'] = df['free_dialogue'].apply(lambda x :''.join(ch.lower() for ch in x if ch not in stoplist))
```

```python
char_corpuses = {}
for char in main_chars:
    char_corpuses[char] = df[df.Character == char]['no_stop_words'].values
```

```python
char_corps = {}
for char, text in char_corpuses.items():
    char_corps[char] = word_tokenize(text)
```

```python
sid_obj = SentimentIntensityAnalyzer()
sentiments = {char: sid_obj.polarity_scores(corpus) for char, corpus in char_corps.items()}
```
